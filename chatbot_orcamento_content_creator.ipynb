```python
#Configura√ß√µes iniciais
import google.generativeai as genai

# Used to securely store your API key
from google.colab import userdata
api_key = userdata.get('SECRET_KEY')

GOOGLE_API_KEY="api_key"
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
# üíñ Defini√ß√£o das vari√°veis do or√ßamento üíñ
valor_base = 199.00
postagem_redes_sociais = 0
uso_imagem_criador = 0
impulsionamento_trafego = 0
qtd_videos_adicionais = 0
qtd_fotos = 0

# üìè Largura fixa para o texto üìè
largura_texto = 80


# ‚ú® Fun√ß√£o para calcular o desconto para v√≠deos adicionais ‚ú®
def calcular_desconto_videos(qtd_videos):
    if qtd_videos == 1:
        return 0.08
    elif qtd_videos == 2:
        return 0.1
    elif qtd_videos >= 3:
        return 0.12
    else:
        return 0


# ‚ú® Fun√ß√£o para calcular o pre√ßo das fotos ‚ú®
def calcular_preco_fotos(qtd_fotos):
    if qtd_fotos == 1:
        return 50.00
    elif qtd_fotos == 2:
        return 80.00
    elif qtd_fotos >= 3:
        return qtd_fotos * 30.00
    else:
        return 0


# ‚ú® Fun√ß√£o para imprimir texto com quebra de linha autom√°tica ‚ú®
def imprimir_texto_formatado(texto):
    palavras = texto.split()
    linha_atual = ""
    for palavra in palavras:
        if len(linha_atual) + len(palavra) + 1 <= largura_texto:
            linha_atual += palavra + " "
        else:
            print(linha_atual.rstrip())
            linha_atual = palavra + " "
    print(linha_atual.rstrip())


# ü§ñ Fun√ß√£o principal do chatbot ü§ñ
def chatbot():
    global postagem_redes_sociais, uso_imagem_criador, \
           impulsionamento_trafego, qtd_videos_adicionais, qtd_fotos

    imprimir_texto_formatado("\n## Monte Seu Or√ßamento de V√≠deo Promocional! üéâ\n")

    # ‚ùì Pergunta 1: Postagem nas Redes Sociais ‚ùì
    imprimir_texto_formatado(
        "Gostaria que eu publicasse o **v√≠deo** üé¨ em minhas **redes sociais**? üòé\n"
        "Tenho seguidores engajados! Isso dar√° mais **visibilidade** "
        "ao seu v√≠deo e atrair√° um **p√∫blico** maior! üöÄ\nDigite 'SIM' para "
        "incluir a **postagem** nas minhas redes sociais, ou 'N√ÉO' para "
        "seguirmos adiante.\n"
    )
    resposta = input()
    if resposta.upper() == "SIM":
        postagem_redes_sociais = 0.2 * valor_base

    print("\n") # Linha em branco

    # ‚ùì Pergunta 2: Uso da Imagem do Criador ‚ùì
    imprimir_texto_formatado(
        "Que tal usar minha **imagem** üì∏ no seu **v√≠deo**? üòâ\n"
        "Minha presen√ßa trar√° mais **credibilidade** e "
        "**confian√ßa** √† sua mensagem, impactando positivamente seu "
        "**p√∫blico**! ü§©\nDigite 'SIM' para incluir minha **imagem** no "
        "v√≠deo, ou 'N√ÉO' para prosseguirmos.\n"
    )
    resposta = input()
    if resposta.upper() == "SIM":
        uso_imagem_criador = 0.2 * valor_base

    print("\n") # Linha em branco

    # ‚ùì Pergunta 3: Impulsionamento de Tr√°fego ‚ùì
    imprimir_texto_formatado(
        "Para alcan√ßar um **p√∫blico** ainda maior e garantir que sua "
        "**mensagem** chegue √†s pessoas certas, recomendo **impulsionar** seu "
        "**v√≠deo** nas **redes sociais**! üì£\nCom um **investimento** "
        "adicional, voc√™ poder√° segmentar o **p√∫blico** e aumentar "
        "significativamente o **alcance** da sua **campanha**! üìà\nDigite "
        "'SIM' para **turbinar** o alcance do seu v√≠deo com o "
        "**impulsionamento**, ou 'N√ÉO' caso prefira seguir com as op√ß√µes "
        "j√° selecionadas.\n"
    )
    resposta = input()
    if resposta.upper() == "SIM":
        impulsionamento_trafego = 0.2 * valor_base

    print("\n") # Linha em branco

    # ‚ùì Pergunta 4: V√≠deos Adicionais ‚ùì
    imprimir_texto_formatado(
        "Para diferentes **plataformas** ou com **mensagens** espec√≠ficas, ter "
        "mais **v√≠deos** pode ser uma √≥tima **estrat√©gia**! üí°\nDeseja "
        "produzir outros **v√≠deos promocionais**, al√©m do primeiro?\nSe sim, "
        "quantos **v√≠deos adicionais** voc√™ gostaria?\nSe n√£o, digite "
        "'0'.\nLembre-se que voc√™ ter√° um **desconto progressivo** no valor "
        "dos v√≠deos adicionais: 8% para 1 v√≠deo extra, 10% para 2 v√≠deos "
        "extras, 12% para 3 v√≠deos extras, e assim por diante!\n"
    )
    resposta = input()
    qtd_videos_adicionais = int(resposta)

    print("\n") # Linha em branco

    # ‚ùì Pergunta 5: Fotos ‚ùì
    imprimir_texto_formatado(
        "Que tal complementar seus **v√≠deos** com **fotos profissionais**? üì∏\n"
        "Elas podem ser utilizadas em suas **redes sociais**, **site** ou outros "
        "**materiais de divulga√ß√£o**.\nQuantas **fotos** voc√™ gostaria de "
        "incluir no pacote?\nSe n√£o quiser fotos, digite '0'.\nLembre-se: 1 "
        "foto custa R$ 50,00, 2 fotos custam R$ 40,00 cada, e 3 fotos ou "
        "mais custam R$ 30,00 cada.\n"
    )
    resposta = input()
    qtd_fotos = int(resposta)

    # üßÆ Calcula o or√ßamento üßÆ
    valor_total = valor_base + postagem_redes_sociais + \
                   uso_imagem_criador + impulsionamento_trafego
    if qtd_videos_adicionais > 0:
        valor_videos_adicionais = (
            valor_base * (qtd_videos_adicionais + 1)
        ) * (1 - calcular_desconto_videos(qtd_videos_adicionais))
        valor_total += valor_videos_adicionais
    valor_fotos = calcular_preco_fotos(qtd_fotos)
    valor_total += valor_fotos

    # üßæ Imprime o or√ßamento detalhado üßæ
    print("\n## Seu Or√ßamento Detalhado: üí∞\n")
    print(f"1. V√≠deo Base: R$ {valor_base:.2f}")
    print("\n2. Servi√ßos Adicionais:")
    if postagem_redes_sociais > 0:
        print(f"    * Postagem nas Minhas Redes Sociais: R$ {postagem_redes_sociais:.2f}")
    if uso_imagem_criador > 0:
        print(f"    * Uso da Minha Imagem: R$ {uso_imagem_criador:.2f}")
    if impulsionamento_trafego > 0:
        print(f"    * Impulsionamento de Tr√°fego: R$ {impulsionamento_trafego:.2f}")

    if qtd_videos_adicionais > 0:
        print(f"\n3. V√≠deos Adicionais ({qtd_videos_adicionais}):")
        print(
            f"    * Valor do Pacote com {qtd_videos_adicionais + 1} "
            f"v√≠deos: R$ {(valor_base * (qtd_videos_adicionais + 1)):.2f}"
        )
        print(
            f"    * Desconto de "
            f"{calcular_desconto_videos(qtd_videos_adicionais) * 100:.0f}% "
            f"para {qtd_videos_adicionais} v√≠deos adicionais: R$ "
            f"{(valor_base * (qtd_videos_adicionais + 1)) * calcular_desconto_videos(qtd_videos_adicionais):.2f}"
        )
        print(f"    * Valor Final dos V√≠deos Adicionais: R$ {valor_videos_adicionais:.2f}")

    if qtd_fotos > 0:
        print(f"\n4. Fotos ({qtd_fotos}):")
        print(f"    * Valor total das fotos: R$ {valor_fotos:.2f}")

    print(f"\n5. Valor Total do Or√ßamento: R$ {valor_total:.2f}")

    imprimir_texto_formatado(
        "\nInvestir em **v√≠deos promocionais** com alta qualidade, **divulga√ß√£o "
        "estrat√©gica** e **fotos profissionais** √© o caminho para **impulsionar** "
        "o sucesso da sua **marca**! üöÄ\n"
    )
    imprimir_texto_formatado("E a√≠, vamos come√ßar a produzir? üòâ\n")


# üöÄ Inicia o chatbot üöÄ
chatbot()
```
