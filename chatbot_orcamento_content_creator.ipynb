```python
#ConfiguraÃ§Ãµes iniciais
import google.generativeai as genai

# Used to securely store your API key
from google.colab import userdata
api_key = userdata.get('SECRET_KEY')

GOOGLE_API_KEY="api_key"
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
# ðŸ’– DefiniÃ§Ã£o das variÃ¡veis do orÃ§amento ðŸ’–
valor_base = 199.00
postagem_redes_sociais = 0
uso_imagem_criador = 0
impulsionamento_trafego = 0
qtd_videos_adicionais = 0
qtd_fotos = 0

# ðŸ“ Largura fixa para o texto ðŸ“
largura_texto = 80


# âœ¨ FunÃ§Ã£o para calcular o desconto para vÃ­deos adicionais âœ¨
def calcular_desconto_videos(qtd_videos):
    if qtd_videos == 1:
        return 0.08
    elif qtd_videos == 2:
        return 0.1
    elif qtd_videos >= 3:
        return 0.12
    else:
        return 0


# âœ¨ FunÃ§Ã£o para calcular o preÃ§o das fotos âœ¨
def calcular_preco_fotos(qtd_fotos):
    if qtd_fotos == 1:
        return 50.00
    elif qtd_fotos == 2:
        return 80.00
    elif qtd_fotos >= 3:
        return qtd_fotos * 30.00
    else:
        return 0


# âœ¨ FunÃ§Ã£o para imprimir texto com quebra de linha automÃ¡tica âœ¨
def imprimir_texto_formatado(texto):
    palavras = texto.split()
    linha_atual = ""
    for palavra in palavras:
        if len(linha_atual) + len(palavra) + 1 <= largura_texto:
            linha_atual += palavra + " "
        else:
            print(linha_atual.rstrip())
            linha_atual = palavra + " "
    print(linha_atual.rstrip())


# ðŸ¤– FunÃ§Ã£o principal do chatbot ðŸ¤–
def chatbot():
    global postagem_redes_sociais, uso_imagem_criador, \
           impulsionamento_trafego, qtd_videos_adicionais, qtd_fotos

    imprimir_texto_formatado("\n## Monte Seu OrÃ§amento de VÃ­deo Promocional! ðŸŽ‰\n")

    # â“ Pergunta 1: Postagem nas Redes Sociais â“
    imprimir_texto_formatado(
        "Gostaria que eu publicasse o **vÃ­deo** ðŸŽ¬ em minhas **redes sociais**? ðŸ˜Ž\n"
        "Tenho seguidores engajados! Isso darÃ¡ mais **visibilidade** "
        "ao seu vÃ­deo e atrairÃ¡ um **pÃºblico** maior! ðŸš€\nDigite 'SIM' para "
        "incluir a **postagem** nas minhas redes sociais, ou 'NÃƒO' para "
        "seguirmos adiante.\n"
    )
    resposta = input()
    if resposta.upper() == "SIM":
        postagem_redes_sociais = 0.2 * valor_base

    print("\n") # Linha em branco

    # â“ Pergunta 2: Uso da Imagem do Criador â“
    imprimir_texto_formatado(
        "Que tal usar minha **imagem** ðŸ“¸ no seu **vÃ­deo**? ðŸ˜‰\n"
        "Minha presenÃ§a trarÃ¡ mais **credibilidade** e "
        "**confianÃ§a** Ã  sua mensagem, impactando positivamente seu "
        "**pÃºblico**! ðŸ¤©\nDigite 'SIM' para incluir minha **imagem** no "
        "vÃ­deo, ou 'NÃƒO' para prosseguirmos.\n"
    )
    resposta = input()
    if resposta.upper() == "SIM":
        uso_imagem_criador = 0.2 * valor_base

    print("\n") # Linha em branco

    # â“ Pergunta 3: Impulsionamento de TrÃ¡fego â“
    imprimir_texto_formatado(
        "Para alcanÃ§ar um **pÃºblico** ainda maior e garantir que sua "
        "**mensagem** chegue Ã s pessoas certas, recomendo **impulsionar** seu "
        "**vÃ­deo** nas **redes sociais**! ðŸ“£\nCom um **investimento** "
        "adicional, vocÃª poderÃ¡ segmentar o **pÃºblico** e aumentar "
        "significativamente o **alcance** da sua **campanha**! ðŸ“ˆ\nDigite "
        "'SIM' para **turbinar** o alcance do seu vÃ­deo com o "
        "**impulsionamento**, ou 'NÃƒO' caso prefira seguir com as opÃ§Ãµes "
        "jÃ¡ selecionadas.\n"
    )
    resposta = input()
    if resposta.upper() == "SIM":
        impulsionamento_trafego = 0.2 * valor_base

    print("\n") # Linha em branco

    # â“ Pergunta 4: VÃ­deos Adicionais â“
    imprimir_texto_formatado(
        "Para diferentes **plataformas** ou com **mensagens** especÃ­ficas, ter "
        "mais **vÃ­deos** pode ser uma Ã³tima **estratÃ©gia**! ðŸ’¡\nDeseja "
        "produzir outros **vÃ­deos promocionais**, alÃ©m do primeiro?\nSe sim, "
        "quantos **vÃ­deos adicionais** vocÃª gostaria?\nSe nÃ£o, digite "
        "'0'.\nLembre-se que vocÃª terÃ¡ um **desconto progressivo** no valor "
        "dos vÃ­deos adicionais: 8% para 1 vÃ­deo extra, 10% para 2 vÃ­deos "
        "extras, 12% para 3 vÃ­deos extras, e assim por diante!\n"
    )
    resposta = input()
    qtd_videos_adicionais = int(resposta)

    print("\n") # Linha em branco

    # â“ Pergunta 5: Fotos â“
    imprimir_texto_formatado(
        "Que tal complementar seus **vÃ­deos** com **fotos profissionais**? ðŸ“¸\n"
        "Elas podem ser utilizadas em suas **redes sociais**, **site** ou outros "
        "**materiais de divulgaÃ§Ã£o**.\nQuantas **fotos** vocÃª gostaria de "
        "incluir no pacote?\nSe nÃ£o quiser fotos, digite '0'.\nLembre-se: 1 "
        "foto custa R$ 50,00, 2 fotos custam R$ 40,00 cada, e 3 fotos ou "
        "mais custam R$ 30,00 cada.\n"
    )
    resposta = input()
    qtd_fotos = int(resposta)

    # ðŸ§® Calcula o orÃ§amento ðŸ§®
    valor_total = valor_base + postagem_redes_sociais + \
                   uso_imagem_criador + impulsionamento_trafego
    if qtd_videos_adicionais > 0:
        valor_videos_adicionais = (
            valor_base * (qtd_videos_adicionais + 1)
        ) * (1 - calcular_desconto_videos(qtd_videos_adicionais))
        valor_total += valor_videos_adicionais
    valor_fotos = calcular_preco_fotos(qtd_fotos)
    valor_total += valor_fotos

    # ðŸ§¾ Imprime o orÃ§amento detalhado ðŸ§¾
    print("\n## Seu OrÃ§amento Detalhado: ðŸ’°\n")
    print(f"1. VÃ­deo Base: R$ {valor_base:.2f}")
    print("\n2. ServiÃ§os Adicionais:")
    if postagem_redes_sociais > 0:
        print(f"    * Postagem nas Minhas Redes Sociais: R$ {postagem_redes_sociais:.2f}")
    if uso_imagem_criador > 0:
        print(f"    * Uso da Minha Imagem: R$ {uso_imagem_criador:.2f}")
    if impulsionamento_trafego > 0:
        print(f"    * Impulsionamento de TrÃ¡fego: R$ {impulsionamento_trafego:.2f}")

    if qtd_videos_adicionais > 0:
        print(f"\n3. VÃ­deos Adicionais ({qtd_videos_adicionais}):")
        print(
            f"    * Valor do Pacote com {qtd_videos_adicionais + 1} "
            f"vÃ­deos: R$ {(valor_base * (qtd_videos_adicionais + 1)):.2f}"
        )
        print(
            f"    * Desconto de "
            f"{calcular_desconto_videos(qtd_videos_adicionais) * 100:.0f}% "
            f"para {qtd_videos_adicionais} vÃ­deos adicionais: R$ "
            f"{(valor_base * (qtd_videos_adicionais + 1)) * calcular_desconto_videos(qtd_videos_adicionais):.2f}"
        )
        print(f"    * Valor Final dos VÃ­deos Adicionais: R$ {valor_videos_adicionais:.2f}")

    if qtd_fotos > 0:
        print(f"\n4. Fotos ({qtd_fotos}):")
        print(f"    * Valor total das fotos: R$ {valor_fotos:.2f}")

    print(f"\n5. Valor Total do OrÃ§amento: R$ {valor_total:.2f}")

    imprimir_texto_formatado(
        "\nInvestir em **vÃ­deos promocionais** com alta qualidade, **divulgaÃ§Ã£o "
        "estratÃ©gica** e **fotos profissionais** Ã© o caminho para **impulsionar** "
        "o sucesso da sua **marca**! ðŸš€\n"
    )
    imprimir_texto_formatado("E aÃ­, vamos comeÃ§ar a produzir? ðŸ˜‰\n")


# ðŸš€ Inicia o chatbot ðŸš€
chatbot()
```
